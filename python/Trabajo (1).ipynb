#### 7.- Realiza una función que mediante la introducción de una secuencia de ADN devuelva lo siguiente: (3pto)
1. Todos los ORFs posibles que puedan surgir de la secuencia.
2. Selecciona los 6 ORFs más largos e indica si corresponden a la strand positiva o negativa.
3. Devuelve como mejor predicción, el más largo de todos.
4. Calcula el porcentaje de aminoácidos aromáticos de esa proteína elegida y su longitud relativa a la secuencia original de ADN.

```python
#apartado 1
import re

#ORFs de la hebra positiva
pos_inicio_positiva = []  #Lista para guardar posiciones de inicio
for m in re.finditer("ATG", secuencia): 
    pos1 = m.start()
    pos_inicio_positiva.append(pos1)

pos_stop_positiva = []   #Lista para guardar posiciones de stop
for m in re.finditer("TAA|TAG|TGA", secuencia): 
    pos2 = m.start()
    pos_stop_positiva.append(pos2)

orfs_positiva = []
for inicio in pos_inicio_positiva: 
    for stop in pos_stop_positiva: 
        if stop > inicio and (stop - inicio) %3 == 0:  #Para asignar a cada inicio su stop
            orf_positiva= secuencia[inicio:stop+3]
            orfs_positiva.append(orf_positiva)
            break
print("Número de ORFS en la hebra positiva:", len(orfs_positiva))
for num, orf in enumerate(orfs_positiva, start=1): 
    print("ORF", num, ":", orf, "Longitud ORF", num, ":", len(orf_positiva))


#ORFs de la hebra negativa
#Primero creo la secuencia complementaria a la hebra positiva
Hebra_negativa = ''
for base in secuencia: 
    if base == "A": 
        Hebra_negativa = Hebra_negativa + "T"
    elif base == "T": 
        Hebra_negativa = Hebra_negativa + "A"
    elif base == "G": 
        Hebra_negativa = Hebra_negativa + "C"
    elif base == "C": 
        Hebra_negativa = Hebra_negativa + "G"
Hebra_negativa = Hebra_negativa[::-1]

pos_inicio_negativa = []    #Lista para guardar posiciones de inicio
for m in re.finditer("ATG", Hebra_negativa): 
    posA = m.start()
    pos_inicio_negativa.append(posA)

pos_stop_negativa = []    #Lista para guardar posiciones de stop
for m in re.finditer("TAA|TAG|TGA", Hebra_negativa): 
    posB = m.start()
    pos_stop_negativa.append(posB)

orfs_negativa = []     
for inicio in pos_inicio_negativa: 
    for stop in pos_stop_negativa: 
        if stop > inicio and (stop - inicio) %3 == 0:    #Para asignar a cada inicio su stop
            orf_negativa= Hebra_negativa[inicio:stop+3]
            orfs_negativa.append(orf_negativa)
            break
print()
print()
print("Número de ORFS en la hebra negativa:", len(orfs_negativa))
for num, orf in enumerate(orfs_negativa, start=1): 
    print("ORF", num, ":", orf, "Longitud ORF", num, ":", len(orf_negativa))
```

```python
#apartado 2
orfs_totales = orfs_negativa + orfs_positiva
longitudes_orfs = []
for orf in orfs_totales:  #Para comprobar a qué hebra pertenecen
    if orf in orfs_positiva: 
        hebra = 'positiva'
    else: 
        hebra = 'negativa'
    longitudes_orfs.append((len(orf), orf, hebra))
longitudes_orfs.sort()
#La función .sort ordena de menor a mayor automáticamente. Como queremos las 6 OFRs más largas, invierto la lista para que las mayores estén las primeras. 
longitudes_orfs = longitudes_orfs[::-1]
for num in range(6):
    longitud, orf, hebra = longitudes_orfs[num]
    print("ORF", num+1, ":", "Longitud:", longitud, "Hebra:", hebra)
```
